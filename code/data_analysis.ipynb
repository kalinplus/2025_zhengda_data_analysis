# 导入模块和数据

```python
import pandas as pd
import csv

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
import numpy
import scipy
import statsmodels

plt.rcParams['font.sans-serif'] = ['SimHei']  # 例如使用 SimHei (黑体)
plt.rcParams['axes.unicode_minus'] = False  # 解决负号 '-' 显示为方块的问题
```

```python
excel_path = '../data/全部问卷信息.xlsx'
df_buyers = pd.read_excel(excel_path, sheet_name='买家重编码')
df_sellers = pd.read_excel(excel_path, sheet_name='卖家重编码')
df_unused = pd.read_excel(excel_path, sheet_name='未使用重编码')

print("买家数据形状: ", df_buyers.shape)
print("卖家数据形状: ", df_sellers.shape)
print("未使用二手平台的受调查者数据形状: ", df_unused.shape)
print("------------------------------------------------------------------")

display("买家数据前5行: ", df_buyers.head())
print("------------------------------------------------------------------")
display("卖家数据前5行: ", df_sellers.head())
print("------------------------------------------------------------------")
display("未使用二手平台的受调查者数据前5行: ", df_unused.head())
```

# 数据预处理

## 案例：买家

主要目的是删除一些无用的列，并重命名一些列

```python
df_buyers = (
    df_buyers[
        ["所用时间", "1、您的性别为：", "2、您的年龄为：", "3、您的学历为：", "4、您目前从事的职业为：", "5、您的月收入区间为：",
         "6、您是否使用过二手交易平台？"]].rename(
        columns={"所用时间": "用时",
                 "1、您的性别为：": "性别",
                 "2、您的年龄为：": "年龄",
                 "3、您的学历为：": "学历",
                 "4、您目前从事的职业为：": "职业",
                 "5、您的月收入区间为：": "收入",
                 "6、您是否使用过二手交易平台？": "是否使用过二手交易平台"
                 }
    ).dropna()
)

# 用时单位默认是秒，转成int类型便于画图
df_buyers.iloc[:, 0] = df_buyers.iloc[:, 0].str[:-1].astype(int)
df_buyers
```

```python
sns.histplot(data=df_buyers, x="用时", bins="auto")
plt.title("买家问卷用时分布")
plt.xlabel("用时（秒）")
plt.ylabel("频数")

threshold_95 = df_buyers["用时"].quantile(0.95)
threshold_99 = df_buyers["用时"].quantile(0.99)

plt.axvline(threshold_95, color='r', linestyle='--', label=f'95%阈值: {threshold_95}')
plt.axvline(threshold_99, color='g', linestyle='--', label=f'99%阈值: {threshold_99}')

plt.show()

print("95%的买家问卷用时少于: ", threshold_95)
print("99%的买家问卷用时少于: ", threshold_99)
```

## 其他身份
将上述预处理整理为方法


```python
def preprocess(df: pd.DataFrame) -> pd.DataFrame:
    df = (
        df[["所用时间", "1、您的性别为：", "2、您的年龄为：", "3、您的学历为：", "4、您目前从事的职业为：",
            "5、您的月收入区间为：", "6、您是否使用过二手交易平台？"]].rename(
            columns={"所用时间": "用时",
                     "1、您的性别为：": "性别",
                     "2、您的年龄为：": "年龄",
                     "3、您的学历为：": "学历",
                     "4、您目前从事的职业为：": "职业",
                     "5、您的月收入区间为：": "收入",
                     "6、您是否使用过二手交易平台？": "是否使用过二手交易平台"
                     }
        ).dropna()
    )
    # 用时单位默认是秒，转成int类型便于画图
    df.iloc[:, 0] = df.iloc[:, 0].str[:-1].astype(int)
    return df


def plot_hist(df: pd.DataFrame, identity: str) -> None:
    sns.histplot(data=df, x="用时", bins="auto")
    plt.title(f"{identity}问卷用时分布")
    plt.xlabel("用时（秒）")
    plt.ylabel("频数")

    threshold_95 = df["用时"].quantile(0.95)
    threshold_99 = df["用时"].quantile(0.99)

    plt.axvline(threshold_95, color='r', linestyle='--', label=f'95%阈值: {threshold_95}')
    plt.axvline(threshold_99, color='g', linestyle='--', label=f'99%阈值: {threshold_99}')

    plt.show()

    print(f"95%的{identity}问卷用时少于: ", threshold_95)
    print(f"99%的{identity}问卷用时少于: ", threshold_99)
```

### 卖家

```python
df_sellers = preprocess(df_sellers)

df_sellers
```

```python
plot_hist(df_sellers, "卖家")
```

### 未使用

```python
df_unused = preprocess(df_unused)
df_unused
```

```python
plot_hist(df_unused, "未使用")
```

## 处理结果

```python
buyers_threshold_99 = df_buyers["用时"].quantile(0.99)
sellers_threshold_99 = df_sellers["用时"].quantile(0.99)
unused_threshold_99 = df_unused["用时"].quantile(0.99)

df_buyers = df_buyers[df_buyers["用时"] < buyers_threshold_99]
df_sellers = df_sellers[df_sellers["用时"] < sellers_threshold_99]
df_unused = df_unused[df_unused["用时"] < unused_threshold_99]

df_buyers = df_buyers.rename(columns={"是否使用过二手交易平台" : "身份"})
df_sellers = df_sellers.rename(columns={"是否使用过二手交易平台" : "身份"})
df_unused = df_unused.rename(columns={"是否使用过二手交易平台" : "身份"})

# 编码身份，0为买家，1为卖家，2为未使用
df_buyers["身份"] = 0
df_sellers["身份"] = 1
df_unused["身份"] = 2

display(df_buyers.head())
display(df_sellers.head())
display(df_unused.head())
```

## 整理结果到一张表
将上述人口统计信息和身份信息整合到一张表中

```python
df_participants = pd.concat([df_buyers, df_sellers, df_unused])

df_participants = pd.get_dummies(df_participants, columns=["身份"], prefix="身份")
df_participants.rename(columns={"身份_0": "买家", "身份_1": "卖家", "身份_2": "未使用"}, inplace=True)

display(df_participants)

# 保存到Excel（包含所有身份数据）
df_participants.to_excel('../result/清除时间异常的数据.xlsx', index=False)
```

# EDA

## 人口统计数据分布

```python
df_participants = df_participants.sort_values(by='年龄', ascending=True)

plt.figure(figsize=(20, 10))

# 选择需要展示的字段（可根据需要调整列表）
columns = ['性别', '年龄', '学历', '职业', '收入']

# enumerate中，index永远在第一个
for idx, col in enumerate(columns, 1):
    plt.subplot(2, 3, idx)  # 2行3列布局
    sns.countplot(data=df_participants, x=col)
    plt.title(f'{col}分布')
    plt.xlabel('')  

plt.tight_layout()
plt.show()
```
